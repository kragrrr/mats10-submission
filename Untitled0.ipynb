```python
import torch
from huggingface_hub import hf_hub_download
import numpy as np

print("="*80)
print("LOADING SAE FOR GEMMA-2-2B LAYER 20")
print("="*80)

# Download SAE weights from the correct repository
print(f"\n[1/3] Downloading SAE weights for layer 20...")

# Correct path for Gemma Scope SAEs
sae_path = hf_hub_download(
    repo_id="google/gemma-scope-2b-pt-res",
    filename="layer_20/width_16k/average_l0_71/params.npz",
    repo_type="model"
)

print(f"✓ Downloaded to: {sae_path}")

# Load SAE weights
print(f"\n[2/3] Loading SAE parameters...")

sae_params = np.load(sae_path)

# Convert to PyTorch tensors
W_enc = torch.from_numpy(sae_params['W_enc']).float().cuda()
W_dec = torch.from_numpy(sae_params['W_dec']).float().cuda()
b_enc = torch.from_numpy(sae_params['b_enc']).float().cuda()
b_dec = torch.from_numpy(sae_params['b_dec']).float().cuda()

print(f"✓ Loaded SAE weights:")
print(f"  W_enc: {W_enc.shape}")
print(f"  W_dec: {W_dec.shape}")
print(f"  b_enc: {b_enc.shape}")
print(f"  b_dec: {b_dec.shape}")

# Create SAE class
print(f"\n[3/3] Creating SAE...")

class SimpleGemmaScope:
    def __init__(self, W_enc, W_dec, b_enc, b_dec):
        self.W_enc = W_enc
        self.W_dec = W_dec
        self.b_enc = b_enc
        self.b_dec = b_dec

    def encode(self, x):
        """Encode layer 20 activations to SAE features"""
        x = x.float().cuda()
        pre_acts = x @ self.W_enc + self.b_enc
        return torch.relu(pre_acts)

    def decode(self, f):
        """Decode SAE features back to activations"""
        return f @ self.W_dec + self.b_dec

sae = SimpleGemmaScope(W_enc, W_dec, b_enc, b_dec)

print(f"✓ SAE ready!")
print(f"  Input dim: {W_enc.shape[0]}")
print(f"  SAE features: {W_enc.shape[1]}")

print("\n" + "="*80)
print("✅ SAE LOADED SUCCESSFULLY")
print("="*80)
```

```python
import torch
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss
import pandas as pd
from datasets import load_dataset

print("="*80)
print("STEP 3: EXTRACT DATA & TRAIN 100 PROBES")
print("="*80)

# ============================================================================
# EXTRACT DATA
# ============================================================================
print("\n[STEP 3A] Extracting embeddings and SAE features from WikiText...")

all_embeddings = []
all_sae_features = []
samples_collected = 0
MAX_SAMPLES = 10000

dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")

for idx, example in enumerate(dataset):
    if samples_collected >= MAX_SAMPLES:
        break

    text = example['text']
    if len(text.strip()) < 20:
        continue

    try:
        inputs = tokenizer(text[:1024], return_tensors='pt', max_length=256,
                          truncation=True, padding='max_length')
        inputs = {k: v.to('cuda') for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model(**inputs, output_hidden_states=True)

            # Layer 0 (embeddings)
            layer0 = outputs.hidden_states[0]
            # Layer 20 (late layer)
            # TODO check the hidden state indexing is correct
            layer20 = outputs.hidden_states[20]

            batch_size, seq_len, hidden_dim = layer20.shape

            # Flatten
            layer0_flat = layer0.reshape(-1, hidden_dim).cpu()
            layer20_flat = layer20.reshape(-1, hidden_dim)

            # Encode with SAE
            sae_acts = sae.encode(layer20_flat).cpu()

            all_embeddings.append(layer0_flat)
            all_sae_features.append(sae_acts)

            samples_collected += seq_len

        if (idx + 1) % 50 == 0:
            print(f"  [{idx+1:4d} docs] {samples_collected:,} samples collected")

    except Exception as e:
        print(f"  Skipped doc {idx}: {str(e)[:40]}")
        continue

# Concatenate
embeddings = torch.cat(all_embeddings, dim=0)
sae_features = torch.cat(all_sae_features, dim=0)

print(f"\n✓ Data extracted:")
print(f"  Embeddings: {embeddings.shape}")
print(f"  SAE features: {sae_features.shape}")

# ============================================================================
# SELECT 100 RANDOM FEATURES
# ============================================================================
print("\n[STEP 3B] Selecting 100 random features...")

n_total = sae_features.shape[1]
random_features = np.random.choice(n_total, size=100, replace=False)

print(f"Selected from {n_total:,} total features")

# ============================================================================
# TRAIN 100 PROBES
# ============================================================================
print("\n[STEP 3C] Training 100 individual probes...\n")

results = []
sample_sizes = []  # Track sample sizes

for probe_idx, feat_idx in enumerate(random_features):
    target_acts = sae_features[:, feat_idx]

    # Check if feature has variation
    n_active = (target_acts > 0.1).sum().item()
    n_total = len(target_acts)

    if n_active == 0 or n_active == n_total:
        continue

    # Get active/inactive indices
    active_idx = torch.where(target_acts > 0.1)[0]
    inactive_idx = torch.where(target_acts <= 0.1)[0]

    n_samples = min(len(active_idx), len(inactive_idx))
    if n_samples < 10:
        continue

    # ✅ TODO 1: Track sample size
    sample_sizes.append({
        'Feature': feat_idx,
        'n_samples': n_samples,
        'n_active': len(active_idx),
        'n_inactive': len(inactive_idx),
    })

    # Split
    n_train = int(n_samples * 0.8)

    # Shuffle the training data indices
    active_perm = active_idx[torch.randperm(len(active_idx))]
    inactive_perm = inactive_idx[torch.randperm(len(inactive_idx))]

    train_idx = torch.cat([active_perm[:n_train], inactive_perm[:n_train]])
    val_idx = torch.cat([active_perm[n_train:], inactive_perm[n_train:]])

    if len(train_idx) == 0 or len(val_idx) == 0:
        continue

    # Get data with ORIGINAL threshold (0.1)
    X_train = embeddings[train_idx].numpy()
    y_train = (target_acts[train_idx] > 0.1).float().numpy()
    X_val = embeddings[val_idx].numpy()
    y_val = (target_acts[val_idx] > 0.1).float().numpy()

    # Train
    clf = LogisticRegression(C=10, max_iter=2000, solver='lbfgs', random_state=42)
    clf.fit(X_train, y_train)

    # Evaluate at ORIGINAL threshold (0.1)
    train_acc = clf.score(X_train, y_train)
    val_acc = clf.score(X_val, y_val)

    # ✅ TODO 2: Test different activation thresholds
    thresholds = [0.01, 0.05, 0.1, 0.5, 1.0]
    threshold_accs = {}

    for threshold in thresholds:
        y_train_th = (target_acts[train_idx] > threshold).float().numpy()
        y_val_th = (target_acts[val_idx] > threshold).float().numpy()

        # Only evaluate if threshold has both classes
        if len(np.unique(y_train_th)) == 2 and len(np.unique(y_val_th)) == 2:
            val_acc_th = clf.score(X_val, y_val_th)
            threshold_accs[threshold] = val_acc_th

    results.append({
        'Feature': feat_idx,
        'Train Acc': train_acc,
        'Val Acc': val_acc,
        'Gap': train_acc - val_acc,
        'n_samples': n_samples,
        'threshold_accs': threshold_accs,  # Store threshold analysis
    })

    if (probe_idx + 1) % 20 == 0:
        print(f"  [{probe_idx+1:3d}/100] Feature {feat_idx}: Train={train_acc:.4f}, Val={val_acc:.4f}, Samples={n_samples}")

# ============================================================================
# ANALYSIS: Sample Sizes & Threshold Effects
# ============================================================================

print("\n" + "="*80)
print("SAMPLE SIZE ANALYSIS")
print("="*80)

sample_df = pd.DataFrame(sample_sizes)
print(f"\nSample Size Statistics:")
print(f"  • Mean samples per feature: {sample_df['n_samples'].mean():.0f}")
print(f"  • Median samples per feature: {sample_df['n_samples'].median():.0f}")
print(f"  • Min: {sample_df['n_samples'].min():.0f}")
print(f"  • Max: {sample_df['n_samples'].max():.0f}")
print(f"  • Mean active ratio: {(sample_df['n_active'] / (sample_df['n_active'] + sample_df['n_inactive'])).mean():.2%}")

# ============================================================================
# THRESHOLD SENSITIVITY ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("THRESHOLD SENSITIVITY ANALYSIS")
print("="*80)

# Collect threshold accuracies
threshold_sensitivity = {0.01: [], 0.05: [], 0.1: [], 0.5: [], 1.0: []}

for result in results:
    for threshold, acc in result['threshold_accs'].items():
        threshold_sensitivity[threshold].append(acc)

print(f"\nValidation Accuracy at Different Thresholds:")
for threshold in [0.01, 0.05, 0.1, 0.5, 1.0]:
    if threshold_sensitivity[threshold]:
        accs = np.array(threshold_sensitivity[threshold])
        print(f"  • Threshold={threshold}: Mean={np.mean(accs):.4f}, Median={np.median(accs):.4f}, N={len(accs)}")
    else:
        print(f"  • Threshold={threshold}: No valid samples")

# Visualize threshold sensitivity
fig, ax = plt.subplots(figsize=(10, 6))

thresholds = [0.01, 0.05, 0.1, 0.5, 1.0]
means = []
stds = []

for threshold in thresholds:
    if threshold_sensitivity[threshold]:
        accs = np.array(threshold_sensitivity[threshold])
        means.append(np.mean(accs))
        stds.append(np.std(accs))
    else:
        means.append(None)
        stds.append(None)

ax.plot(thresholds, means, marker='o', markersize=8, linewidth=2.5, color='steelblue', label='Mean Accuracy')
ax.fill_between(thresholds,
                 [m - s if m is not None else None for m, s in zip(means, stds)],
                 [m + s if m is not None else None for m, s in zip(means, stds)],
                 alpha=0.3, color='steelblue', label='±1 Std Dev')
ax.set_xlabel('Activation Threshold', fontsize=11, fontweight='bold')
ax.set_ylabel('Validation Accuracy', fontsize=11, fontweight='bold')
ax.set_title('Probe Accuracy Across Different Activation Thresholds\n(Is it harder to predict weaker activations?)',
             fontsize=12, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_xscale('log')

plt.tight_layout()
plt.savefig('threshold_sensitivity_analysis.png', dpi=150, bbox_inches='tight')
print(f"\n✓ Saved: threshold_sensitivity_analysis.png")
plt.show()

print("\n" + "="*80)
print("KEY INSIGHT")
print("="*80)
print(f"""
If accuracy DROPS at higher thresholds (e.g., 0.1 vs 1.0):
  → Easier to predict WEAK activations (close to 0.1)
  → Harder to predict EXTREME activations (>1.0)
  → Suggests embeddings encode "active/inactive" boundary, not magnitude

If accuracy STABLE across thresholds:
  → Embeddings encode magnitude information too
  → Linear probe can predict across full activation range
""")

# ============================================================================
# RESULTS
# ============================================================================
print("\n" + "="*80)
print("FINAL RESULTS: 100 RANDOM FEATURES")
print("="*80)

df = pd.DataFrame(results)
accs = df['Val Acc'].values
gaps = df['Gap'].values

print(f"\nProbes trained: {len(df)}/100")
print(f"\n📊 VALIDATION ACCURACY:")
print(f"  Mean:   {accs.mean():.4f}")
print(f"  Median: {np.median(accs):.4f}")
print(f"  Std:    {accs.std():.4f}")
print(f"  Min:    {accs.min():.4f}")
print(f"  Max:    {accs.max():.4f}")

print(f"\n📊 OVERFITTING GAP (Train - Val):")
print(f"  Mean:   {gaps.mean():.4f}")
print(f"  Median: {np.median(gaps):.4f}")

print(f"\n📊 ACCURACY DISTRIBUTION:")
print(f"  ≥95%: {(accs >= 0.95).sum()}/{len(accs)}")
print(f"  ≥90%: {(accs >= 0.90).sum()}/{len(accs)}")
print(f"  ≥85%: {(accs >= 0.85).sum()}/{len(accs)}")
print(f"  ≥80%: {(accs >= 0.80).sum()}/{len(accs)}")
print(f"  ≥75%: {(accs >= 0.75).sum()}/{len(accs)}")

print("\n" + "="*80)
print("✅ EXPERIMENT COMPLETE!")
print("="*80)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
from tqdm import tqdm
from transformers import AutoModelForCausalLM, AutoTokenizer

print("="*80)
print("CROSS-LAYER TRANSCODER + RESIDUAL SAE PIPELINE")
print("="*80)

# ============================================================================
# STEP 0: Load Model and Extract Layer 20 for All Tokens
# ============================================================================

print("\n" + "="*80)
print("STEP 0: LOADING MODEL AND EXTRACTING LAYER 20")
print("="*80)

# Load Gemma-2-2B model and tokenizer
print("\nLoading Gemma-2-2B model...")
model_name = "google/gemma-2-2b"
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map='auto'
)
model.eval()
print(f"✓ Model loaded on {device}")

# Extract layer 20 activations
all_layer20 = []
samples_collected = 0
MAX_SAMPLES = 10240

print(f"\nExtracting layer 20 activations...")
dataset_for_l20 = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")

for idx, example in enumerate(dataset_for_l20):
    if samples_collected >= MAX_SAMPLES:
        break

    text = example['text']
    if len(text.strip()) < 20:
        continue

    try:
        inputs = tokenizer(text[:1024], return_tensors='pt', max_length=256,
                          truncation=True, padding='max_length')
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model(**inputs, output_hidden_states=True)
            layer20 = outputs.hidden_states[20]  # [1, 256, 2304]

        batch_size, seq_len, hidden_dim = layer20.shape
        layer20_flat = layer20.reshape(-1, hidden_dim).cpu().float()

        all_layer20.append(layer20_flat)
        samples_collected += seq_len

        if (idx + 1) % 50 == 0:
            print(f"  [{idx+1:4d} docs] {samples_collected:,} samples collected")

    except Exception as e:
        print(f"  Error on doc {idx}: {str(e)[:50]}")
        continue

# Check if we collected any data
if len(all_layer20) == 0:
    raise ValueError("No data collected! Check model loading and dataset access.")

layer20_full = torch.cat(all_layer20, dim=0)[:MAX_SAMPLES]
print(f"\n✓ Layer 20 extracted: {layer20_full.shape}")

# Also extract embeddings (layer 0) for the transcoder
print(f"\nExtracting layer 0 embeddings...")
all_embeddings = []
samples_collected = 0

for idx, example in enumerate(dataset_for_l20):
    if samples_collected >= MAX_SAMPLES:
        break

    text = example['text']
    if len(text.strip()) < 20:
        continue

    try:
        inputs = tokenizer(text[:1024], return_tensors='pt', max_length=256,
                          truncation=True, padding='max_length')
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model(**inputs, output_hidden_states=True)
            layer0 = outputs.hidden_states[0]  # [1, 256, 2304]

        batch_size, seq_len, hidden_dim = layer0.shape
        layer0_flat = layer0.reshape(-1, hidden_dim).cpu().float()

        all_embeddings.append(layer0_flat)
        samples_collected += seq_len

        if (idx + 1) % 50 == 0:
            print(f"  [{idx+1:4d} docs] {samples_collected:,} samples collected")

    except Exception as e:
        continue

if len(all_embeddings) == 0:
    raise ValueError("No embeddings collected!")

embeddings = torch.cat(all_embeddings, dim=0)[:MAX_SAMPLES]
print(f"\n✓ Embeddings extracted: {embeddings.shape}")

print("\n" + "="*80)
print("✓ DATA EXTRACTION COMPLETE")
print("="*80)
print(f"  • Embeddings (layer 0): {embeddings.shape}")
print(f"  • Layer 20: {layer20_full.shape}")
print(f"\nReady for transcoder training!")
```

```python
# ============================================================================
# STEP 1: TRAINING CROSS-LAYER TRANSCODER
# ============================================================================

print("\n" + "="*80)
print("STEP 1: TRAINING CROSS-LAYER TRANSCODER")
print("="*80)

# class CrossLayerTranscoder(nn.Module):
#     def __init__(self, hidden_dim=2304, bottleneck_dim=4096):
#         super().__init__()
#         self.encoder = nn.Sequential(
#             nn.Linear(hidden_dim, bottleneck_dim),
#             nn.ReLU(),
#             nn.Linear(bottleneck_dim, bottleneck_dim),
#             nn.ReLU(),
#         )
#         self.decoder = nn.Sequential(
#             nn.Linear(bottleneck_dim, bottleneck_dim),
#             nn.ReLU(),
#             nn.Linear(bottleneck_dim, hidden_dim),
#         )


    # def forward(self, x):
    #     z = self.encoder(x)
    #     return self.decoder(z)

# Affine Map Encoder
class CrossLayerTranscoder(nn.Module):
    def __init__(self, hidden_dim=2304):
        super().__init__()
        # Simple affine transformation: y = Wx + b
        self.transform = nn.Linear(hidden_dim, hidden_dim)

    def forward(self, x):
        return self.transform(x)



device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
transcoder = CrossLayerTranscoder().to(device)
optimizer = optim.Adam(transcoder.parameters(), lr=1e-4)
criterion = nn.MSELoss()

# ============================================================================
# NUCLEAR OPTION: Force complete detachment via numpy conversion
# ============================================================================
print(f"\nPreparing data (nuclear detachment)...")
print(f"  Layer 0 embeddings: {embeddings.shape}")
print(f"  Layer 20 activations: {layer20_full.shape}")

# Step 1: Force everything to CPU numpy (breaks ALL grad connections)
emb_np = embeddings.detach().cpu().numpy() if torch.is_tensor(embeddings) else np.array(embeddings)
l20_np = layer20_full.detach().cpu().numpy() if torch.is_tensor(layer20_full) else np.array(layer20_full)

# Step 2: Create completely fresh tensors (NO grad history possible)
embeddings_clean = torch.tensor(emb_np, dtype=torch.float32, requires_grad=False).to(device)
layer20_clean = torch.tensor(l20_np, dtype=torch.float32, requires_grad=False).to(device)

print(f"  ✓ Cleaned embeddings: {embeddings_clean.shape}, requires_grad={embeddings_clean.requires_grad}")
print(f"  ✓ Cleaned layer20: {layer20_clean.shape}, requires_grad={layer20_clean.requires_grad}")

# Create dataset
dataset = TensorDataset(embeddings_clean, layer20_clean)
batch_size = 256
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Train
print(f"\nTraining transcoder for 20 epochs...")
transcoder.train()

# Enable gradients for training (they were disabled globally in Cell 2)
torch.set_grad_enabled(True)

for epoch in range(20):
    total_loss = 0
    num_batches = 0

    for batch_embeddings, batch_layer20 in dataloader:
        optimizer.zero_grad()

        predictions = transcoder(batch_embeddings)
        loss = criterion(predictions, batch_layer20)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        num_batches += 1

    avg_loss = total_loss / num_batches
    print(f"  Epoch {epoch+1:2d}/20: Loss = {avg_loss:.6f}")

# Disable gradients again after training to save memory
torch.set_grad_enabled(False)

print(f"\n✓ Transcoder trained!")

# Evaluate
print(f"\nTranscoder Performance:")
transcoder.eval()
with torch.no_grad():
    all_predictions = transcoder(embeddings_clean)
    mse = criterion(all_predictions, layer20_clean).item()

    ss_res = torch.sum((layer20_clean - all_predictions) ** 2).item()
    ss_tot = torch.sum((layer20_clean - layer20_clean.mean()) ** 2).item()
    r2 = 1 - (ss_res / ss_tot)

    print(f"  MSE: {mse:.6f}")
    print(f"  R²:  {r2:.4f}")
    print(f"  Explains {r2*100:.1f}% of variance in layer 20")
```

```python
# ============================================================================
# STEP 2: COMPUTE RESIDUALS (EMERGENT INFORMATION)
# ============================================================================

print("\n" + "="*80)
print("STEP 2: COMPUTING RESIDUALS")
print("="*80)

print(f"\nComputing: Layer20 - Transcoder(Layer0)")

transcoder.eval()
with torch.no_grad():
    layer20_predictions = transcoder(embeddings_clean)
    residuals_clean = layer20_clean - layer20_predictions

print(f"\nResidual Statistics:")
print(f"  Shape: {residuals_clean.shape}")
print(f"  Mean: {residuals_clean.mean():.6f}")
print(f"  Std:  {residuals_clean.std():.6f}")
print(f"  Min:  {residuals_clean.min():.6f}")
print(f"  Max:  {residuals_clean.max():.6f}")

layer20_var = layer20_clean.var().item()
residual_var = residuals_clean.var().item()

print(f"\nVariance Partition:")
print(f"  Layer 20 variance:           {layer20_var:.6f}")
print(f"  Residual variance:           {residual_var:.6f}")
print(f"  % explained by transcoder:   {(1 - residual_var/layer20_var)*100:.1f}%")
print(f"  % unexplained (emergent):    {(residual_var/layer20_var)*100:.1f}%")

print(f"\n✓ Residuals computed!")
```

```python
# ============================================================================
# STEP 3A: TRAIN BASELINE SAE (on original layer 20)
# ============================================================================

print("\n" + "="*80)
print("STEP 3A: TRAINING BASELINE SAE (on layer 20 directly)")
print("="*80)

class SimpleAutoencoder(nn.Module):
    def __init__(self, input_dim=2304, hidden_dim=16384):
        super().__init__()
        self.encoder = nn.Linear(input_dim, hidden_dim)
        self.decoder = nn.Linear(hidden_dim, input_dim)

    def encode(self, x):
        return torch.relu(self.encoder(x))

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        z = self.encode(x)
        return self.decode(z)

print(f"\nTraining baseline SAE on layer 20...")
sae_baseline = SimpleAutoencoder().to(device)
optimizer_baseline = optim.Adam(sae_baseline.parameters(), lr=1e-3)

sparsity_weight = 0.01

# ✅ FIX: Create a version of layer20_clean that can be used for training
# The issue is layer20_clean has requires_grad=False
# We need to allow gradients through the SAE, not through the data
layer20_data = layer20_clean.detach()  # Detach once, use throughout

# ✅ FIX: Enable gradients for training (they were disabled globally in Cell 2)
torch.set_grad_enabled(True)
sae_baseline.train()

for epoch in range(50):
    total_loss = 0

    for i in range(0, len(layer20_data), 256):
        batch = layer20_data[i:i+256]

        optimizer_baseline.zero_grad()

        # Forward pass through SAE
        z = sae_baseline.encode(batch)
        recon = sae_baseline.decode(z)

        # Compute losses
        recon_loss = nn.MSELoss()(recon, batch)
        sparsity_loss = sparsity_weight * torch.mean(torch.abs(z))

        loss = recon_loss + sparsity_loss

        # Backward pass
        loss.backward()
        optimizer_baseline.step()

        total_loss += loss.item()

    if (epoch + 1) % 10 == 0:
        print(f"  Epoch {epoch+1:2d}/50: Loss={total_loss:.6f}")

# Disable gradients again after training to save memory
torch.set_grad_enabled(False)

print(f"\n✓ Baseline SAE trained!")

```

```python
# ============================================================================
# STEP 3B: TRAIN RESIDUAL SAE (on emergent information)
# ============================================================================

print("\n" + "="*80)
print("STEP 3B: TRAINING RESIDUAL SAE (on emergent information)")
print("="*80)

print(f"\nTraining residual SAE on residuals...")
sae_residual = SimpleAutoencoder().to(device)
optimizer_residual = optim.Adam(sae_residual.parameters(), lr=1e-3)

# ✅ FIX: Enable gradients for training (they were disabled globally in Cell 2)
torch.set_grad_enabled(True)
sae_residual.train()

# ✅ FIX: Ensure residuals are detached (no grad history)
residuals_data = residuals_clean.detach()

for epoch in range(50):
    total_loss = 0

    for i in range(0, len(residuals_data), 256):
        batch = residuals_data[i:i+256]

        optimizer_residual.zero_grad()

        z = sae_residual.encode(batch)
        recon = sae_residual.decode(z)

        recon_loss = nn.MSELoss()(recon, batch)
        sparsity_loss = sparsity_weight * torch.mean(torch.abs(z))

        loss = recon_loss + sparsity_loss
        loss.backward()
        optimizer_residual.step()

        total_loss += loss.item()

    if (epoch + 1) % 10 == 0:
        print(f"  Epoch {epoch+1:2d}/50: Loss={total_loss:.6f}")

# Disable gradients again after training to save memory
torch.set_grad_enabled(False)

print(f"\n✓ Residual SAE trained!")

print("\n" + "="*80)
print("✓ PIPELINE STEP 1-3B COMPLETE!")
print("="*80)
print(f"""
KEY FINDINGS SO FAR:
  • Transcoder explains {r2*100:.1f}% of layer 20
  • {(residual_var/layer20_var)*100:.1f}% is emergent information
  • Baseline SAE now ready for probing
  • Residual SAE now ready for probing

NEXT: Run Steps 4-7 to test feature predictability
""")
```

```python
# ============================================================================
# STEP 4: EXTRACT FEATURES FROM BOTH SAES
# ============================================================================

print("\n" + "="*80)
print("STEP 4: EXTRACTING FEATURES FROM BOTH SAES")
print("="*80)

with torch.no_grad():
    baseline_features = sae_baseline.encode(layer20_clean).cpu().numpy()
    residual_features = sae_residual.encode(residuals_clean).cpu().numpy()

print(f"\nBaseline SAE features: {baseline_features.shape}")
print(f"Residual SAE features: {residual_features.shape}")
print(f"\n✓ Features extracted!")
```

```python
# ============================================================================
# STEP 5: TEST BASELINE FEATURES PREDICTABILITY
# ============================================================================

print("\n" + "="*80)
print("STEP 5: TESTING BASELINE SAE FEATURES PREDICTABILITY")
print("="*80)

print(f"\nTesting {min(100, baseline_features.shape[1])} baseline features...")

baseline_results = []
n_test_baseline = min(100, baseline_features.shape[1])
embeddings_np = embeddings_clean.cpu().numpy()

for feat_idx in np.random.choice(baseline_features.shape[1], n_test_baseline, replace=False):
    target = baseline_features[:, feat_idx]

    n_active = (target > 0.1).sum()
    if n_active == 0 or n_active == len(target):
        continue

    active_idx = np.where(target > 0.1)[0]
    inactive_idx = np.where(target <= 0.1)[0]

    n_samples = min(len(active_idx), len(inactive_idx))
    if n_samples < 10:
        continue

    n_train = int(n_samples * 0.8)

    train_idx = np.concatenate([
        np.random.choice(active_idx, n_train, replace=False),
        np.random.choice(inactive_idx, n_train, replace=False)
    ])
    val_idx = np.concatenate([
        np.setdiff1d(active_idx, train_idx[:n_train]),
        np.setdiff1d(inactive_idx, train_idx[n_train:])
    ])

    X_train = embeddings_np[train_idx]
    y_train = (baseline_features[train_idx, feat_idx] > 0.1).astype(int)
    X_val = embeddings_np[val_idx]
    y_val = (baseline_features[val_idx, feat_idx] > 0.1).astype(int)

    if len(np.unique(y_train)) < 2 or len(np.unique(y_val)) < 2:
        continue

    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, y_train)

    val_acc = clf.score(X_val, y_val)
    baseline_results.append({'Feature': feat_idx, 'Val Acc': val_acc})

baseline_results_df = pd.DataFrame(baseline_results)

print(f"\n{'='*60}")
print(f"BASELINE SAE FEATURES: PREDICTABILITY FROM LAYER 0")
print(f"{'='*60}")

if len(baseline_results_df) > 0:
    baseline_accs = baseline_results_df['Val Acc'].values
    print(f"\nFeatures tested: {len(baseline_accs)}")
    print(f"Mean accuracy: {baseline_accs.mean():.4f}")
    print(f"Median accuracy: {np.median(baseline_accs):.4f}")
    print(f"Std dev: {baseline_accs.std():.4f}")
    print(f"\nAccuracy tiers:")
    print(f"  ≥90%: {(baseline_accs >= 0.90).sum()}/{len(baseline_accs)} ({(baseline_accs >= 0.90).sum()/len(baseline_accs)*100:.1f}%)")
    print(f"  ≥85%: {(baseline_accs >= 0.85).sum()}/{len(baseline_accs)} ({(baseline_accs >= 0.85).sum()/len(baseline_accs)*100:.1f}%)")
    print(f"  ≥80%: {(baseline_accs >= 0.80).sum()}/{len(baseline_accs)} ({(baseline_accs >= 0.80).sum()/len(baseline_accs)*100:.1f}%)")
    print(f"  <60%: {(baseline_accs < 0.60).sum()}/{len(baseline_accs)} ({(baseline_accs < 0.60).sum()/len(baseline_accs)*100:.1f}%)")
```

```python
# ============================================================================
# STEP 6: TEST RESIDUAL FEATURES PREDICTABILITY
# ============================================================================

print("\n" + "="*80)
print("STEP 6: TESTING RESIDUAL SAE FEATURES PREDICTABILITY")
print("="*80)

print(f"\nTesting {min(100, residual_features.shape[1])} residual features...")

residual_results = []
n_test_residual = min(100, residual_features.shape[1])

for feat_idx in np.random.choice(residual_features.shape[1], n_test_residual, replace=False):
    target = residual_features[:, feat_idx]

    n_active = (target > 0.1).sum()
    if n_active == 0 or n_active == len(target):
        continue

    active_idx = np.where(target > 0.1)[0]
    inactive_idx = np.where(target <= 0.1)[0]

    n_samples = min(len(active_idx), len(inactive_idx))
    if n_samples < 10:
        continue

    n_train = int(n_samples * 0.8)

    train_idx = np.concatenate([
        np.random.choice(active_idx, n_train, replace=False),
        np.random.choice(inactive_idx, n_train, replace=False)
    ])
    val_idx = np.concatenate([
        np.setdiff1d(active_idx, train_idx[:n_train]),
        np.setdiff1d(inactive_idx, train_idx[n_train:])
    ])

    X_train = embeddings_np[train_idx]
    y_train = (residual_features[train_idx, feat_idx] > 0.1).astype(int)
    X_val = embeddings_np[val_idx]
    y_val = (residual_features[val_idx, feat_idx] > 0.1).astype(int)

    if len(np.unique(y_train)) < 2 or len(np.unique(y_val)) < 2:
        continue

    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, y_train)

    val_acc = clf.score(X_val, y_val)
    residual_results.append({'Feature': feat_idx, 'Val Acc': val_acc})

residual_results_df = pd.DataFrame(residual_results)

print(f"\n{'='*60}")
print(f"RESIDUAL SAE FEATURES: PREDICTABILITY FROM LAYER 0")
print(f"{'='*60}")

if len(residual_results_df) > 0:
    residual_accs = residual_results_df['Val Acc'].values
    print(f"\nFeatures tested: {len(residual_accs)}")
    print(f"Mean accuracy: {residual_accs.mean():.4f}")
    print(f"Median accuracy: {np.median(residual_accs):.4f}")
    print(f"Std dev: {residual_accs.std():.4f}")
    print(f"\nAccuracy tiers:")
    print(f"  ≥90%: {(residual_accs >= 0.90).sum()}/{len(residual_accs)} ({(residual_accs >= 0.90).sum()/len(residual_accs)*100:.1f}%)")
    print(f"  ≥85%: {(residual_accs >= 0.85).sum()}/{len(residual_accs)} ({(residual_accs >= 0.85).sum()/len(residual_accs)*100:.1f}%)")
    print(f"  ≥80%: {(residual_accs >= 0.80).sum()}/{len(residual_accs)} ({(residual_accs >= 0.80).sum()/len(residual_accs)*100:.1f}%)")
    print(f"  <60%: {(residual_accs < 0.60).sum()}/{len(residual_accs)} ({(residual_accs < 0.60).sum()/len(residual_accs)*100:.1f}%)")
```
